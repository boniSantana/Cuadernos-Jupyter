```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftfreq
```

```python
Ts = 10e-4
Fs = 1/Ts
Tf = 0.1
N = 100
```

```python
# Generaremos los valores del dominio de una forma distinta a como se hace en Mathlab, usaremos numpy.linspace.
# np.linspace ( start, stop, num, endpoint=true), genera num muestras entre start y stop, se incluye stop si endpoint = true.

t = np.linspace (0, Tf, N)
```

```python
# Si quiero ver que tiene t descomento y ejecuto la siguiente línea:
#print (t)
```

```python
f0=50;
w0=2*np.pi*f0;
```

```python
# Genero los valores de la imagen.

y = 220*np.sqrt(2)*np.cos(w0*t);
```

```python
# Ploteo

plt.plot(t,y)
```

```python
# Podemos configurar el gráfico, por ejemplo:
plt.xlabel('Tiempo (s)')
plt.ylabel('y(t)')
plt.plot (t,y, 'bo', alpha=0.5, color='xkcd:lightish blue')
plt.plot (t,y, 'k', alpha=1, color='xkcd:dusk')
```

# ¡Hora de transformar!

```python
#  Para el dominio de la función usamos fftfreq
distanceTwoPoints= t[1]-t[0]
f2 = fftfreq(N, distanceTwoPoints)
```

```python
y_transform = fft(y)
```

```python
# Podemos configurar el gráfico, por ejemplo:
plt.xlabel('frecuencia [$\omega$]')
plt.ylabel('y_transformada(t)')
plt.plot (f2,abs(y_transform), 'bo', alpha=0.5, color='xkcd:lightish blue')
plt.plot (f2,abs(y_transform), 'k', alpha=1, color='xkcd:dusk')
```

```python
# Podemos configurar el gráfico, por ejemplo:
plt.xlabel('frecuencia [$\omega$]')
plt.ylabel('y_transformada(t)')
plt.stem (f2,abs(y_transform), use_line_collection='true')
```
